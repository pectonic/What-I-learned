```python
y = dataFrame['Type'].values
x = dataFrame.drop('Type',axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain,xTest,yTrain,yTest = train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(xTrain)
```

```python
xTrain = scaler.transform(xTrain)
```

```python
xTest = scaler.transform(xTest)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping # overfitting cozumu
```

```python
xTrain.shape
```

```python
model = Sequential()


model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=15,activation='relu'))
model.add(Dense(units=15,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer = 'adam')
```

```python
model.fit(x=xTrain,y=yTrain, epochs=700,validation_data=(xTest,yTest),verbose=1)
```

```python
model.history.history 
```

```python
modelLoss= pd.DataFrame(model.history.history)
```

```python
modelLoss.plot() # overfitting alert
```

```python
model = Sequential()


model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=15,activation='relu'))
model.add(Dense(units=15,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer = 'adam')
```

```python
earlyStop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=xTrain,y=yTrain, epochs=700,validation_data=(xTest,yTest),verbose=1, callbacks=[earlyStop])
```

```python
modelLoss = pd.DataFrame(model.history.history)
```

```python
modelLoss.plot()
```

```python
model = Sequential()


model.add(Dense(units=30,activation='relu'))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation='relu'))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation='relu'))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer = 'adam')
```

```python
model.fit(x=xTrain,y=yTrain, epochs=700,validation_data=(xTest,yTest),verbose=1, callbacks=[earlyStop])
```

```python
modelLoss = pd.DataFrame(model.history.history)
```

```python
modelLoss.plot()
```

```python
tahminler = (model.predict(xTest) > 0.5).astype("int32")
```

```python
tahminler
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(yTest,tahminler)) # %91 dogruluk 
```

```python
print(confusion_matrix(yTest,tahminler)) # 6 deger yanlis digerleri dogru
```
