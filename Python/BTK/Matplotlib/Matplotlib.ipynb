# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1 - Giris

```python
yasListesi = np.array([10,20,30,30,30,40,50,60,70,75])
kiloListesi = np.array([20,60,80,85,86,87,70,90,95,90])

plt.plot(yasListesi,kiloListesi,"g") # g: green
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun yasa orani")
plt.show() # normalde grafik ciktisini bununla aliriz ama %matplotlib inline bize kolaylik saglayarak kendisi olusturuyor
```

## 2 - customize

```python
dizi1 = np.linspace(0,10,20)
dizi1
```

```python
dizi1Kup = dizi1 ** 7
dizi1Kup
```

```python
dizi1Kare = dizi1 ** 2
dizi1Kare
```

```python
plt.plot(dizi1,dizi1Kup, "b*-") # b--
```

```python
plt.subplot(1,2,1)
plt.plot(dizi1, dizi1Kup, "r*-")
plt.subplot(1,2,2) # 1 sira olacak, 2 column olacak ve su an ben 2. grafigi ciziyorum
plt.plot(dizi1Kup, dizi1, "g--")
```

```python
firstFigure = plt.figure()
figureAxex = firstFigure.add_axes([0.1,0.1,1.9,0.9])
figureAxex.plot(dizi1, dizi1Kup,"g")
figureAxex.set_xlabel("X Eksen veri ismi")
figureAxex.set_ylabel("Y Eksen veri ismi")
figureAxex.set_title("Grafik Basligi")
```

## 3 -  e Tablolar

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7]) 
eksen2 = figure2.add_axes([0.2,0.35,0.3,0.3]) # ic tablo bu belirledik, pozisyonlari vs.

eksen1.plot(dizi1,dizi1Kup,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("Ana Grafik Baslik")


eksen2.plot(dizi1Kup,dizi1,"b")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("Sub Grafik Baslik")
```

```python
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(dizi1,dizi1Kup,"b")
```

## 4 - Subplots

```python
(benimFigure, benimEksen) = plt.subplots(nrows=1, ncols=2)
type(benimEksen)
for eksen in benimEksen:
    eksen.plot(dizi1,dizi1Kup,"b")
    eksen.set_xlabel("X eksen")
plt.tight_layout() # grafiklerin araligini duzeltir
#benimFigure
```

## 5 - visualization

```python
newFigure = plt.figure(figsize=(6,6),dpi=100) # dpi: pixel reso, figsize: figure size

newAxis = newFigure.add_axes([0.1,0.1,0.9,0.9])

newAxis.plot(dizi1, dizi1 ** 2, label = "dizi1 ** 2") # label grafike ne oldugunu yazar
newAxis.plot(dizi1, dizi1 ** 3, label = "dizi1 ** 3") 
newAxis.legend(loc=6) # grafigin labelleri gostermeye yarar ve loc: location loc=1/2/3/4/5/6 vs
```

```python
newFigure.savefig("myFigure.png",dpi=300)
```

## 6 - advanced visualization

```python
(myFigure, myAxis) = plt.subplots()
myAxis.plot(dizi1, dizi1Kare,"b", alpha=0.3) # alpha saydamliktir
myAxis.plot(dizi1Kare, dizi1, color="#1ce1ce") # hex color kullanilabilir OYYY RENGE BAK LA 
```

```python
(newFigure,newAxis) = plt.subplots()

newAxis.plot(dizi1, dizi1+2, "b", linewidth = 5.2)
newAxis.plot(dizi1, dizi1+4, "r")

newAxis.plot(dizi1, dizi1, "g", linestyle="--")
newAxis.plot(dizi1, dizi1-2, "purple", linestyle=":")
newAxis.plot(dizi1, dizi1+8, "orange", linestyle="-.")

newAxis.plot(dizi1, dizi1-5, "black", linestyle="--", marker="o", markersize = 6 ,markerfacecolor="#1ce1ce")
newAxis.plot(dizi1, dizi1-10, "black", linestyle="--", marker="+", markersize = 12)

```
