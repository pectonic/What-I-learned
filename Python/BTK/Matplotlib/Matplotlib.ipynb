# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1 - Giris

```python
yasListesi = np.array([10,20,30,30,30,40,50,60,70,75])
kiloListesi = np.array([20,60,80,85,86,87,70,90,95,90])

plt.plot(yasListesi,kiloListesi,"g") # g: green
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun yasa orani")
plt.show() # normalde grafik ciktisini bununla aliriz ama %matplotlib inline bize kolaylik saglayarak kendisi olusturuyor
```

## 2 - customize

```python
dizi1 = np.linspace(0,10,20)
dizi1
```

```python
dizi1Kup = dizi1 ** 7
dizi1Kup
```

```python
plt.plot(dizi1,dizi1Kup, "b*-") # b--
```

```python
plt.subplot(1,2,1)
plt.plot(dizi1, dizi1Kup, "r*-")
plt.subplot(1,2,2) # 1 sira olacak, 2 column olacak ve su an ben 2. grafigi ciziyorum
plt.plot(dizi1Kup, dizi1, "g--")
```

```python
firstFigure = plt.figure()
figureAxex = firstFigure.add_axes([0.1,0.1,1.9,0.9])
figureAxex.plot(dizi1, dizi1Kup,"g")
figureAxex.set_xlabel("X Eksen veri ismi")
figureAxex.set_ylabel("Y Eksen veri ismi")
figureAxex.set_title("Grafik Basligi")
```

## 3 -  e Tablolar

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7]) 
eksen2 = figure2.add_axes([0.2,0.35,0.3,0.3]) # ic tablo bu belirledik, pozisyonlari vs.

eksen1.plot(dizi1,dizi1Kup,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("Ana Grafik Baslik")


eksen2.plot(dizi1Kup,dizi1,"b")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("Sub Grafik Baslik")
```

```python
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(dizi1,dizi1Kup,"b")
```
